```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
X = np.random.rand(100,1)*10
true_b = 2
true_w = 3
y = true_w * X + true_b + np.random.randn(100, 1)

w = 0
b = 0
lr = 0.01
epochs = 1000
for i in range(epochs):
  y_pred = w * X + b
  dw = -2 * np.mean(X * (y - y_pred))
  db = -2 * np.mean(y - y_pred)
  w -= lr * dw
  b -= lr * db

print(f"\nFinal model: y = {w:.2f}x + {b:.2f}")


plt.scatter(X, y, label="Data")
plt.plot(X, w * X + b, color="red", label="Learned Line")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Linear Regression from Scratch")
plt.legend()
plt.grid(True)
plt.show()
```
